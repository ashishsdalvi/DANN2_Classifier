## Create Variant Df 

```python
def generate_variant_df(vcf_file):
    """
    Converts vcf file to pandas dataframe and filters to ONLY keep SNV's
    """
    rows = []
    
    with gzip.open(vcf_file, "r") as f:
        for line in tqdm(f):
            line = line.decode("utf-8") # decode 

            if line.startswith("#"): # skip headers 
                continue
            
            variant = line.split("\t")
            chrom = variant[0]
            pos = int(variant[1])
            ref = variant[3]
            alt = variant[4]

            info = variant[7].split(";")
            allele_count = info[0].split("=")[1]
            allele_number = info[1].split("=")[1]
            allele_frequency = info[2].split("=")[1]

            # make sure variant is a snv
            if len(ref)==1 and len(alt)==1:
                rows.append({'chrom': chrom, 'start': pos-1, 'end': pos, 'ref': ref, 'alt': alt, 'AC': allele_count, 'AN': allele_number, 'AF': allele_frequency})

            # if len(rows) > 10000:
            #     break

    variant_df = pd.DataFrame(rows)

    return variant_df

def subset_variants(bed_file, vcf_file):
    """
    Returns a pandas df of the subsetted variants 
    """
    regions = bf.read_table(bed_file, schema="bed6")
    regions.columns = ["chrom", "start", "end", "feature_name", "associated_element", "annotation"]
    
    variant_df = generate_variant_df(vcf_file)
  
    overlap = bf.overlap(regions, variant_df, how='inner', suffixes=('_bed', '_var')) # take overlap 
    return overlap 
```

## Create Model Input

```python
def create_model_input_df(variant_df):
    # returns the input for forzoi (a dataframe) and set of labels for common variants (0) and singletons (1)
    # common is >1% and rare is <0.5% in pop. 
    chrom_list, pos_list, ref_list, alt_list = [], [], [], []
    variant_labels = []

    for i, row in variant_df.iterrows():
        af = row['AF_var']
        ac = row['AC_var']
        chrom, pos, ref, alt = row['chrom_var'], row['end_var'], row['ref_var'], row['alt_var']
        # if common variant
        if af > 0.01 and ac > 1:
            chrom_list.append(chrom)
            pos_list.append(pos)
            ref_list.append(ref)
            alt_list.append(alt)
            variant_labels.append(0) # 0 is common
            variant_label = 0
        # if singleton 
        elif ac == 1:
            chrom_list.append(chrom)
            pos_list.append(pos)
            ref_list.append(ref)
            alt_list.append(alt)
            variant_labels.append(1) # 1 is singleton
            variant_label = 1 

    df_variants = pd.DataFrame({
        'chrom': chrom_list,
        'pos': pos_list,
        'ref': ref_list,
        'alt': alt_list,
        'variant_label': variant_labels
    })

    return df_variants
```

```python
import pandas as pd
import duckdb
import tempfile

def fast_duckdb_cadd_merge_memsafe(cadd_tsv_path, variant_csv_path):
    # Load and clean df_variants in memory
    df = pd.read_csv(variant_csv_path)
    if 'Unnamed: 0' in df.columns:
        df = df.drop(columns='Unnamed: 0')

    # Create a temp CSV with no index, cleaned columns
    with tempfile.NamedTemporaryFile(suffix='.csv', mode='w+', delete=False) as tmp:
        temp_csv_path = tmp.name
        df.to_csv(temp_csv_path, index=False)

    # Run DuckDB query
    con = duckdb.connect()
    query = f"""
    SELECT v.chrom, v.pos, v.ref, v.alt, v.variant_label, c.raw_score, c.phred_score
    FROM read_csv('{temp_csv_path}',
                  delim=',',
                  header=True,
                  columns={{
                      'chrom': 'VARCHAR',
                      'pos': 'INT',
                      'ref': 'VARCHAR',
                      'alt': 'VARCHAR',
                      'variant_label': 'INT'
                  }},
                  ignore_errors=true,
                  strict_mode=false
    ) v
    JOIN read_csv('{cadd_tsv_path}',
                  delim='\\t',
                  header=False,
                  columns={{
                      'chrom': 'VARCHAR',
                      'pos': 'INT',
                      'ref': 'VARCHAR',
                      'alt': 'VARCHAR',
                      'raw_score': 'FLOAT',
                      'phred_score': 'FLOAT'
                  }},
                  ignore_errors=true,
                  strict_mode=false
    ) c
    ON replace(v.chrom, 'chr', '') = c.chrom
       AND v.pos = c.pos
       AND v.ref = c.ref
       AND v.alt = c.alt
    """
    result = con.execute(query).df()

    # Optional: delete temp CSV
    import os
    os.remove(temp_csv_path)

    return result
```

```python
def subset_CADD_scores(CADD_scores_tsv, df_variants, chunksize=1_000_000):
    # Add composite key to variants
    df_variants = df_variants.copy()
    df_variants['key'] = (
        df_variants['chrom'].str.replace('chr', '', regex=False) + '_' +
        df_variants['pos'].astype(str) + '_' +
        df_variants['ref'] + '_' +
        df_variants['alt']
    )

    variant_keys = set(df_variants['key'])
    results = []

    col_names = ['chrom', 'pos', 'ref', 'alt', 'phred_score', 'raw_score']

    for chunk in pd.read_csv(CADD_scores_tsv, sep='\t', names=col_names, chunksize=chunksize):
        chunk['key'] = (
            chunk['chrom'].astype(str) + '_' +
            chunk['pos'].astype(str) + '_' +
            chunk['ref'] + '_' +
            chunk['alt']
        )
        # Filter and deduplicate
        filtered_chunk = chunk[chunk['key'].isin(variant_keys)].drop_duplicates('key')
        results.append(filtered_chunk[['key', 'phred_score', 'raw_score']])

    # Combine and merge
    cadd_scores = pd.concat(results, ignore_index=True)
    merged = df_variants.merge(cadd_scores, on='key', how='left')
    merged.drop(columns='key', inplace=True)

    return merged



# def subset_CADD_scores(CADD_scores_tsv, df_variants):
#     # Load CADD scores
#     chr_CADD_scores = pd.read_csv(CADD_scores_tsv, sep="\t", index_col=None, 
#                                     names=['chrom', 'pos', 'ref', 'alt', 'phred_score', 'raw_score'])

#     # Build dictionary for fast lookup
#     cadd_dict = {
#         f"{row.chrom}_{row.pos}_{row.ref}_{row.alt}": (row.phred_score, row.raw_score)
#         for row in chr_CADD_scores.itertuples(index=False)
#     }

#     # Generate keys from df_variants (without modifying original)
#     keys = (
#         df_variants['chrom'].str.replace('chr', '', regex=False) + '_' +
#         df_variants['pos'].astype(str) + '_' +
#         df_variants['ref'] + '_' +
#         df_variants['alt']
#     )

#     # Lookup CADD scores
#     phred_scores, raw_scores = zip(*[
#         cadd_dict.get(k, (float('nan'), float('nan')))
#         for k in keys
#     ])

#     # Construct output DataFrame
#     result = df_variants.copy()
#     result['phred_score'] = phred_scores
#     result['raw_score'] = raw_scores

#     return result
```

## Convert Parquet Files to Hdf5

```python
def parquet_to_h5(parquet_path, h5_path):
    pf = pq.ParquetFile(parquet_path)
    schema = pf.schema_arrow

    numeric_cols = [
        f.name for f in schema
        if patypes.is_integer(f.type) or patypes.is_floating(f.type)
    ]

    with h5py.File(h5_path, 'w') as h5f:
        created = set()

        for rg in range(pf.num_row_groups):
            table = pf.read_row_group(rg, columns=numeric_cols)

            for col in numeric_cols:
                data = table.column(col).to_numpy()

                if col not in created:
                    h5f.create_dataset(
                        name=col,
                        data=data,
                        maxshape=(None,),
                        chunks=True,
                        compression="lzf"  # optional: faster than gzip
                    )
                    created.add(col)
                else:
                    dset = h5f[col]
                    old_size = dset.shape[0]
                    new_size = old_size + len(data)
                    dset.resize((new_size,))
                    dset[old_size:] = data

    print(f"Conversion done: {parquet_path} → {h5_path}")
```
